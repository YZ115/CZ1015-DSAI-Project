# Exploratory Analysis

```python
AE = pd.read_csv('athlete_events.csv')
print("Number of data from all countries: ",AE.shape)
print("\n")
```

```python
Regions = pd.read_csv('noc_regions.csv')
#ID
ID = AE.ID.unique()
ID = pd.DataFrame(ID)
ID['IDNumber']=ID.index
ID.columns=['ID','IDNumber']
#Years
years = AE.Year.unique()
years = pd.DataFrame(years)
years['yearNumber']=years.index
years.columns=['Year','yearNumber']
#Events
events = AE.Event.unique()
events = pd.DataFrame(events)
events['eventNumber']=events.index
events.columns=['Event','eventNumber']
#Sports
sports = AE.Sport.unique()
sports = pd.DataFrame(sports)
sports['sportNumber']=sports.index
sports.columns=['Sport','sportNumber']
#Sex
sex = AE.Sex.unique()
sex = pd.DataFrame(sex)
sex['sexNumber']=sex.index
sex.columns=['Sex','sexNumber']
#teams
teams = AE.Team.unique()
teams = pd.DataFrame(teams)
teams['teamNumber']=teams.index
teams.columns=['Team','teamNumber']
#NOC
noc = AE.NOC.unique()
noc = pd.DataFrame(noc)
noc['nocNumber']=noc.index
noc.columns=['NOC','nocNumber']
#Games
games = AE.Games.unique()
games = pd.DataFrame(games)
games['gamesNumber']=games.index
games.columns=['Games','gamesNumber']
#seasons
seasons = AE.Season.unique()
seasons = pd.DataFrame(seasons)
seasons['seasonNumber']=seasons.index
seasons.columns=['Season','seasonNumber']
#Cities
cities = AE.City.unique()
cities = pd.DataFrame(cities)
cities['cityNumber']=cities.index
cities.columns=['City','cityNumber']
#Names
names = AE.Name.unique()
names = pd.DataFrame(names)
names['nameNumber']=names.index
names.columns=['Name','nameNumber']
#Medals
medals = AE.Medal.unique()
medals = pd.DataFrame(medals)
medals['medalNumber']=medals.index
medals.columns=['Medal','medalNumber']
medals = medals.replace([1,2,3],1)
display_side_by_side(ID.head(),years.head(),events.head(),sports.head(),sex.head(),teams.head(),noc.head(),games.head(),seasons.head(),cities.head(),names.head(),medals.head())
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
AE=AE.interpolate()
AE.head()
```

```python
joint=pd.merge(AE,years,how='left',on='Year')
joint=pd.merge(joint,events,how='left',on='Event')
joint=pd.merge(joint,sports,how='left',on='Sport')
joint=pd.merge(joint,sex,how='left',on='Sex')
joint=pd.merge(joint,teams,how='left',on='Team')
joint=pd.merge(joint,noc,how='left',on='NOC')
joint=pd.merge(joint,games,how='left',on='Games')
joint=pd.merge(joint,seasons,how='left',on='Season')
joint=pd.merge(joint,cities,how='left',on='City')
joint=pd.merge(joint,names,how='left',on='Name')
joint=pd.merge(joint,medals,how='left',on='Medal')
joint=joint[['ID','Name','nameNumber','Sex','sexNumber','Age','Height','Weight','Team','teamNumber','NOC','nocNumber','Games','gamesNumber','Year','yearNumber','Season','seasonNumber','City','cityNumber','Sport','sportNumber','Event','eventNumber','Medal','medalNumber']]
joint=joint.interpolate()
jointNumeric=pd.DataFrame(joint[['ID','sexNumber','Age','Height','Weight','teamNumber','nocNumber','gamesNumber','yearNumber','seasonNumber','cityNumber','sportNumber','eventNumber','medalNumber']])
#joint = joint.loc[(joint['Sport'] == 'Athletics')]
joint.head(1)
```

# Upsampling Minority (Class balancing)

```python
from sklearn.utils import resample
totalMajority=joint_dummies[joint.medalNumber==0]
totalMinority=joint_dummies[joint.medalNumber==1]
print("Majority one-hot:\t",totalMajority.shape,"\n")
print("Minority one-hot:\t",totalMinority.shape,"\n")
```

```python
totalMinority_up = resample(totalMinority,
                            replace=True,     # sample with replacement
                            n_samples=231333, # to match majority class
                            random_state=123) # reproducible results
```

```python
joint_upsampled=pd.concat([totalMajority, totalMinority_up])
joint_upsampled['medalNumber'].value_counts()
```

# Random Forest Classifier

```python
responseMedal = pd.DataFrame(joint_upsampled["medalNumber"])
predictors = pd.DataFrame(joint_upsampled)
predictors = predictors.drop(columns=['medalNumber'])
predictors=predictors.interpolate()
predictors.dropna(inplace=True)
responseMedal = pd.concat([predictors,responseMedal],axis=1,join_axes=[predictors.index])
responseMedal = pd.DataFrame(responseMedal["medalNumber"])
responseMedal.dropna(inplace=True)
responseMedal.info()
predictors.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Split the Dataset into Train and Test
predictortrain, predictortest, responsetrain, responsetest = train_test_split(predictors, responseMedal, test_size = 0.2)

# Check the sample sizes
print("Train Set :", responsetrain.shape, predictortrain.shape)
print("Test Set  :", responsetest.shape, predictortest.shape)
```

```python
#Will most likely take a long time
forestTotal = RandomForestClassifier(n_estimators=10,random_state=0)
forestTotal.fit(predictortrain,responsetrain)
```

```python
tscoretrain=forestTotal.score(predictortrain,responsetrain)
tscoretest=forestTotal.score(predictortest,responsetest)
print("World")
print("Accuracy on train dataset: \t",tscoretrain)
print("Accuracy on test dataset: \t",tscoretest)
```

```python
timportances=forestTotal.feature_importances_
print(timportances)
```

```python
importancesTotal = pd.DataFrame(forestTotal.feature_importances_,
                                   index = predictortrain.columns,
                                    columns=['importance']).sort_values('importance',ascending=False)
```
